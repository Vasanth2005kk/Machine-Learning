### Scatter plot

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size",palette="YlGnBu")
```

### Histograms plot

```python
sns.histplot(tips["tip"],kde=True) # bins=10 add prameter
```

### Distributtion plot

```python
sns.displot(tips["tip"],kde=True,bins=20)
```

### Bar plot

```python
sns.barplot(x="sex",y="tip",data=tips)
```

### Box plot

```python
sns.boxplot(x="day",y="total_bill",data=tips)
sns.set_style("whitegrid")
```

### strip plot

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

### join plot

```python
sns.jointplot(x="tip",y="total_bill",data=tips,kind="reg")
sns.jointplot(x="tip",y="total_bill",data=tips,kind="kde")
sns.jointplot(x="tip",y="total_bill",data=tips,kind="hex")
```

### pair plot

```python
sns.pairplot(titanic.select_dtypes(['number']),hue="pclass")
```

## save file

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example Titanic dataset (replace with your actual DataFrame)
titanic = sns.load_dataset('titanic')

# Ensure 'pclass' is included in the numeric data
numeric_columns = titanic.select_dtypes(['number']).copy()
numeric_columns['pclass'] = titanic['pclass']

# Create the pair plot
sns_plot = sns.pairplot(numeric_columns, hue='pclass')

# Save the plot as an SVG
sns_plot.savefig('pairplot.svg')
```
