```python
# Create a figure with custom size and resolution
plt.figure(figsize=(5, 4), dpi=300)

# Data for the first plot
x = [2, 3, 4, 5, 6]
y = [2, 4, 7, 8, 9]

# First plot with markers and line style
plt.plot(x, y, 'r^--', label='2x')

# Data for the second plot
x2 = np.arange(0, 4.5, 0.5)  # Start: 0, End: 4.5 (exclusive), Step: 0.5
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')  # Solid line for the first part
plt.plot(x2[5:], x2[5:]**2, 'r--')  # Dashed line for the second part

# Title with font customization
plt.title("My First Graph", fontdict={
    "fontname": "cmss10",
    "fontsize": 30
})

# X and Y axis labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show legend
plt.legend()


# Display the plot
plt.show()

plt.savefig('myfrsitGraph.svg',dpi=300)
```

# Bar chart

```python
lable = ["A","B","C"]
values = [2,5,1]

plt.figure(figsize=(5, 4), dpi=70)

bars = plt.bar(lable,values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('o')

patterns=['/','*','o']
for i in bars:
    i.set_hatch(patterns.pop(0))


plt.show()
```
